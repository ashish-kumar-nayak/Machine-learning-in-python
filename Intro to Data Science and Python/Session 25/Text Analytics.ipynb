## Building a Spam Classifier

```python
#https://archive.ics.uci.edu/ml/datasets/sms+spam+collection
```

```python
import pandas as pd
```

```python
# importing the Dataset
messages = pd.read_csv(r'C:\Users\v-ashnay\Documents\Ashish\Machine-learning-in-python\Intro to Data Science and Python\Session 25\smsspamcollection\SMSSpamCollection', sep='\t',
                           names=["label", "message"])
```

```python
messages.head()
```

```python
messages.shape
```

```python
#Data cleaning and preprocessing
corpus1 = []
for i in range(0, len(messages)):
    review = re.sub('[^a-zA-Z]', ' ', messages['message'][i])
    review = review.lower()
    review = review.split()
    #review = [wordnet.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus1.append(review)
```

```python
print(corpus1)
```

```python
# Creating the Bag of Words model
from sklearn.feature_extraction.text import CountVectorizer
spamcv = CountVectorizer(max_features=2500) 
spamX = spamcv.fit_transform(corpus1).toarray()
print(spamX)
```

```python
spamX.shape #
```

```python
y=pd.get_dummies(messages['label'])
y.head()
```

```python
y=y.iloc[:,1].values
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(spamX, y, test_size = 0.20, random_state = 0)
```

```python
# Training model using Naive bayes classifier
from sklearn.naive_bayes import MultinomialNB
spam_detect_model = MultinomialNB().fit(X_train, y_train)

y_pred=spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
accuracy_score(y_test, y_pred)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
from gensim.models import Word2Vec
```

```python
# Preprocessing the data
text = re.sub(r'\[[0-9]*\]',' ',paragraph)
text = re.sub(r'\s+',' ',text)
text = text.lower()
text = re.sub(r'\d',' ',text)
text = re.sub(r'\s+',' ',text)
```

```python
text
```

```python
# Preparing the dataset
sentences = nltk.sent_tokenize(text)
sentences
```

```python
#Tokenisation
sentences = [nltk.word_tokenize(sentence) for sentence in sentences]
sentences
```

```python
#Stopwords Removal
print(stopwords.words('english'))
```

```python
for i in range(len(sentences)):
    sentences[i] = [word for word in sentences[i] if word not in stopwords.words('english')]
```

```python
sentences
```

```python
# Training the Word2Vec model
model = Word2Vec(sentences, min_count=1)
```

```python
words = model.wv.vocab
words
#each and every word is represented in terms of vectors for 32 dimensions
```
