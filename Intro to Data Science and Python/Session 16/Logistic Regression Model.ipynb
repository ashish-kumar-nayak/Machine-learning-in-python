```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import numpy as np
BC_data = load_breast_cancer()
```

```python
y = data_df.Target.values 
x_data = data_df.drop(['Target'], axis = 1) 
```

```python
#Normalisation - Min-Max
x = (x_data - np.min(x_data))/(np.max(x_data) - np.min(x_data)).values 
```

```python
pd.DataFrame(x)
```

```python
#Splitting data for training and testing
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.15, random_state = 42) 
```

```python
print("x train: ", x_train.shape) 
print("x test: ", x_test.shape) 
print("y train: ", y_train.shape) 
print("y test: ", y_test.shape)
```

```python
from sklearn import linear_model 
logreg = linear_model.LogisticRegression(random_state = 1000, max_iter = 150) 
```

```python
print("test accuracy: {} ".format(logreg.fit(x_train, y_train).score(x_test, y_test))) 
print("train accuracy: {} ".format(logreg.fit(x_train, y_train).score(x_train, y_train))) 
```
