### Logical operations

```python
df2[(df2.mpg > 25) | (df2.cyl == 8)]
```

```python
df2[((df2.mpg > 25) & (df2.hp < 150)) | ((df2.cyl >= 8) & (df2.hp.between(140, 150)))]
```
